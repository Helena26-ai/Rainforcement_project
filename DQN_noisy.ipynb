```python
# CELL 1: Importy i konfiguracja
import random
import numpy as np
import typing as tt
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from collections import deque

import gymnasium as gym
import ptan
from ptan.experience import ExperienceFirstLast
from ptan.agent import DQNAgent, TargetNet
from ptan.actions import ArgmaxActionSelector
from ptan.experience import ExperienceReplayBuffer, ExperienceSourceFirstLast

from ignite.engine import Engine, Events
from ignite.metrics import RunningAverage
from ignite.contrib.handlers.tensorboard_logger import TensorboardLogger, OutputHandler

from torchrl.modules import NoisyLinear
```

```python
# CELL 2: Hiperparametry
data_dir = "trained_models"
import os
os.makedirs(data_dir, exist_ok=True)

class Hyperparams:
    def __init__(self):
        self.env_name = "PongNoFrameskip-v4"
        self.stop_reward = 19.0
        self.run_name = "pong"
        self.replay_size = 100_000
        self.replay_initial = 10_000
        self.target_net_sync = 1000
        self.epsilon_frames = 100_000
        self.epsilon_final = 0.02
        self.learning_rate = 7.14e-5
        self.batch_size = 32
        self.gamma = 0.99
        self.episodes_to_solve = 500

params = Hyperparams()
```

```python
# CELL 3: Sieć Noisy DQN
class NoisyDQN(nn.Module):
    def __init__(self, input_shape: tt.Tuple[int, ...], n_actions: int):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(input_shape[0], 32, kernel_size=8, stride=4),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=4, stride=2),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, stride=1),
            nn.ReLU(),
            nn.Flatten(),
        )
        conv_out_size = self.conv(torch.zeros(1, *input_shape)).shape[1]
        self.noisy_layers = [
            NoisyLinear(conv_out_size, 512),
            NoisyLinear(512, n_actions)
        ]
        self.fc = nn.Sequential(
            self.noisy_layers[0],
            nn.ReLU(),
            self.noisy_layers[1],
        )

    def forward(self, x):
        x = x.float() / 255.0
        return self.fc(self.conv(x))

    def reset_noise(self):
        for layer in self.noisy_layers:
            layer.reset_noise()

    @torch.no_grad()
    def noisy_layers_sigma_snr(self) -> tt.List[float]:
        return [
            ((layer.weight_mu ** 2).mean().sqrt() /
             (layer.weight_sigma ** 2).mean().sqrt()).item()
            for layer in self.noisy_layers
        ]
```

```python
# CELL 4: Funkcje pomocnicze

def unpack_batch(batch):
    states, actions, rewards, dones, next_states = [], [], [], [], []
    for exp in batch:
        states.append(exp.state)
        actions.append(exp.action)
        rewards.append(exp.reward)
        dones.append(exp.last_state is None)
        next_states.append(exp.state if exp.last_state is None else exp.last_state)
    return np.array(states), np.array(actions), np.array(rewards), np.array(dones), np.array(next_states)

def calc_loss_dqn(batch, net, tgt_net, gamma, device):
    states, actions, rewards, dones, next_states = unpack_batch(batch)
    states_v = torch.tensor(states).to(device)
    next_states_v = torch.tensor(next_states).to(device)
    actions_v = torch.tensor(actions).to(device)
    rewards_v = torch.tensor(rewards, dtype=torch.float32).to(device)
    done_mask = torch.BoolTensor(dones).to(device)

    state_action_vals = net(states_v).gather(1, actions_v.unsqueeze(-1)).squeeze(-1)
    with torch.no_grad():
        next_state_vals = tgt_net(next_states_v).max(1)[0]
        next_state_vals[done_mask] = 0.0
    expected_vals = rewards_v + gamma * next_state_vals
    return nn.MSELoss()(state_action_vals, expected_vals)
```

```python
# CELL 5: Trening

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
env = ptan.common.wrappers.wrap_dqn(gym.make(params.env_name))
net = NoisyDQN(env.observation_space.shape, env.action_space.n).to(device)
print("🧪 Weryfikacja: parametry NoisyLinear")
for name, param in net.named_parameters():
    if "noisy" in name and not param.requires_grad:
        print(f"🚨 Parametr {name} nie ma requires_grad=True!")
tgt_net = TargetNet(net)
selector = ArgmaxActionSelector()
agent = DQNAgent(net, selector, device)
exp_source = ExperienceSourceFirstLast(env, agent, gamma=params.gamma)
buffer = ExperienceReplayBuffer(exp_source, buffer_size=params.replay_size)
optimizer = optim.Adam(net.parameters(), lr=params.learning_rate)

best_mean_reward = -float('inf')
best_model_path = os.path.join(data_dir, "best_model.dat")
rewards_history = []
reward_iterations = []
snr_1_list = []
snr_2_list = []


def batch_generator(buffer, initial, batch_size):
    buffer.populate(initial)
    while True:
        buffer.populate(1)
        yield buffer.sample(batch_size)


def process_batch(engine, batch):
    optimizer.zero_grad()
    loss_v = calc_loss_dqn(batch, net, tgt_net.target_model, params.gamma, device)
    loss_v.backward()
    optimizer.step()
    net.reset_noise()

    if engine.state.iteration % params.target_net_sync == 0:
        tgt_net.sync()
    if engine.state.iteration % 100 == 0:
        snr = net.noisy_layers_sigma_snr()
        engine.state.metrics['snr_1'] = snr[0]
        engine.state.metrics['snr_2'] = snr[1]
        snr_1_list.append(snr[0])
        snr_2_list.append(snr[1])

    # Dodatkowa diagnostyka co 5000 iteracji
    if engine.state.iteration % 5000 == 0:
        print(f"📊 [Iteracja {engine.state.iteration}]")
        for i, layer in enumerate(net.noisy_layers):
            mu_rms = (layer.weight_mu ** 2).mean().sqrt().item()
            sigma_rms = (layer.weight_sigma ** 2).mean().sqrt().item()
            print(f"  Layer {i+1} — SNR: {snr[i]:.4f} | μ RMS: {mu_rms:.5f} | σ RMS: {sigma_rms:.6f}")
    return {'loss': loss_v.item()}

engine = Engine(process_batch)
RunningAverage(output_transform=lambda out: out['loss']).attach(engine, 'avg_loss')

def episode_completed(engine):
    rewards = exp_source.pop_total_rewards()
    if rewards:
        reward = rewards[-1]
        rewards_history.append(reward)
        reward_iterations.append(engine.state.iteration)  # Zapisujemy iterację końca epizodu
        mean_reward = np.mean(rewards_history[-100:])
        print(f"Episode {engine.state.episode}: reward={reward:.2f}, mean_100={mean_reward:.2f}")

        global best_mean_reward
        if mean_reward > best_mean_reward:
            print(f"New best mean reward: {mean_reward:.2f}, saving model")
            best_mean_reward = mean_reward
            torch.save(net.state_dict(), best_model_path)
        if mean_reward >= params.stop_reward:
            print(f"Solved in {engine.state.episode} episodes!")
            engine.terminate()


engine.add_event_handler(Events.ITERATION_COMPLETED(every=1), lambda e: setattr(e.state, 'episode', len(rewards_history)))
engine.add_event_handler(Events.ITERATION_COMPLETED(every=1), lambda e: episode_completed(e))
engine.run(batch_generator(buffer, params.replay_initial, params.batch_size))
```

```python
# CELL 6: Testowanie i nagrywanie jako MP4
import imageio.v2 as imageio

test_env = ptan.common.wrappers.wrap_dqn(gym.make(params.env_name, render_mode="rgb_array"))
net.load_state_dict(torch.load(best_model_path))
net.eval()

video_frames = []
obs, _ = test_env.reset()
done = False

while not done:
    obs_v = torch.tensor([obs]).to(device)
    q_vals = net(obs_v)
    action = torch.argmax(q_vals, dim=1).item()
    obs, _, terminated, truncated, _ = test_env.step(action)
    done = terminated or truncated
    frame = test_env.render()
    video_frames.append(frame)

# Zapis jako MP4
mp4_path = "pong_test_run.mp4"
with imageio.get_writer(mp4_path, fps=30, codec='libx264') as writer:
    for frame in video_frames:
        writer.append_data(frame)

print(f"✅ Zapisano nagranie testu agenta do pliku: {mp4_path}")
```

```python
# CELL 7: Rysowanie wykresów

# Wykres nagrody per epizod (teraz: per iteracja)
plt.figure(figsize=(12, 5))
plt.plot(reward_iterations, rewards_history)
plt.title("Nagroda w funkcji iteracji")
plt.xlabel("Iteracja")
plt.ylabel("Nagroda")
plt.grid(True)

# Średnia z ostatnich 100 nagród (także po iteracjach)
moving_avg = [np.mean(rewards_history[max(0, i - 99):i + 1]) for i in range(len(rewards_history))]
plt.figure(figsize=(12, 5))
plt.plot(reward_iterations, moving_avg)
plt.title("Średnia nagroda (100) w funkcji iteracji")
plt.xlabel("Iteracja")
plt.ylabel("Średnia nagroda (100)")
plt.grid(True)

# Wykres SNR
plt.figure(figsize=(12, 5))
plt.plot([i * 100 for i in range(len(snr_1_list))], snr_1_list, label="SNR Layer 1")
plt.plot([i * 100 for i in range(len(snr_2_list))], snr_2_list, label="SNR Layer 2")
plt.title("SNR w warstwach Noisy (co 100 iteracji)")
plt.xlabel("Iteracja")
plt.ylabel("SNR")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
